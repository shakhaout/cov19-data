```python
import pandas as pd
import numpy as np
import re
import lxml

from bs4 import BeautifulSoup
from requests import get
%matplotlib inline
from datetime import datetime,date
```

```python
#url= 'https://www.somoynews.tv/coronavirus/bangladesh'
url = 'http://corona.gov.bd/'
```

```python
page = get(url)
soup = BeautifulSoup(page.content, 'lxml') 
```

```python
print(soup.prettify())
```

```python
match = soup.title.text
print(match)
```

```python
match_daily = soup.find('ul', class_="statistic__efected-new")
print(match_daily)
```

```python
d_count=[]
for num in match_daily.find_all('li'):
    count = num.text
    d_count.append(count)
del d_count[0]
d_count
```

```python
daily_count=[]
for i in range(len(d_count)):
    dc=d_count[i].replace('(','')
    dc=dc.replace('+','')
    dc=dc.replace(')','')
    dc=int(dc)
    daily_count.append(dc)

l_update = datetime.now()
daily_count.append(l_update)
daily_count
```

```python
df_d = pd.DataFrame(daily_count,index=['Confirmed','Recovered','Death','Last_Update'])
DF_d = df_d.T
DF_d
```

```python
file_name=date.today()
```

```python
DF_d.to_csv('{}.csv'.format(file_name),  encoding='utf-8')
```
